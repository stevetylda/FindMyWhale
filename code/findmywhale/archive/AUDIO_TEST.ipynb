```python
# ---------------------------------- #
#           IMPORT MODULES           #

# General Modules
import io
import os
import pandas as pd
import itertools
import subprocess

# Data Modules
import quilt3 as q3
import boto3
from botocore import UNSIGNED
from botocore.config import Config
from botocore.exceptions import ClientError
import ffmpeg

# Modeling Libraries
import librosa
from pydub import AudioSegment

# Visualization Libraries
import plotly.express as px

#                                    #
# ---------------------------------- #

# ---------------------------------- #
#              FUNCTIONS             #





def preprocess_s3_metadata(data):

    # Update Columns
    data.columns = [str.upper(i) for i in data.columns]

    # Subset Columns
    data = data[["KEY", "ETAG", "SIZE", "LASTMODIFIED"]]

    # Select Only Object with TS Tag
    data = data[data.KEY.str.endswith(".ts")]

    # Get Location Name
    data["LOCATION"] = data["KEY"].str.split("/").str[0]

    # Get Unix - Date
    data["UNIX_DATETIME"] = data["KEY"].str.split("/").str[-2]
    data["UNIX_DATETIME"] = data["UNIX_DATETIME"].astype(int)

    # Convert to Datetime
    data["DATETIME"] = pd.to_datetime(data["UNIX_DATETIME"], unit="s", utc=True)
    data["DATETIME"] = data["DATETIME"].dt.tz_localize(None)

    # Remove Time Stamp
    data["LASTMODIFIED"] = data["LASTMODIFIED"].dt.tz_localize(None)

    return data

def generate_presigned_url(bucket_name, object_name, expiration=3600):
    """
    Generates a presigned URL to retrieve an S3 object.

    :param bucket_name: The name of the S3 bucket.
    :param object_name: The key (path and filename) of the object in the bucket.
    :param expiration: The number of seconds the presigned URL is valid for.
    :return: The presigned URL as a string, or None if an error occurs.
    """
    s3_client = boto3.client("s3", config=Config(signature_version=UNSIGNED))
    try:
        url = s3_client.generate_presigned_url(
            ClientMethod="get_object",
            Params={"Bucket": bucket_name, "Key": object_name},
            ExpiresIn=expiration,
        )
    except ClientError as e:
        print(f"Error generating presigned URL: {e}")
        return None
    return url


#                                    #
# ---------------------------------- #
```

```python
unix_datetime = 1539203407
location_ = "rpi_bush_point"
```

```python
# Filter Data To Build Audio File
DT_HLS_METADATA_FILT = DT_HLS_METADATA[
    (DT_HLS_METADATA.LOCATION == location_)
    & (DT_HLS_METADATA.UNIX_DATETIME == unix_datetime)
].copy()
DT_HLS_METADATA_FILT["KEY_ID"] = DT_HLS_METADATA_FILT["KEY"].apply(
    lambda x: int(x.split("live")[-1].split(".")[0]) + 1
)
DT_HLS_METADATA_FILT = DT_HLS_METADATA_FILT.sort_values("KEY_ID").reset_index(drop=True)
```

```python
DT_HLS_METADATA_FILT["PRESIGNED_LINK"] = DT_HLS_METADATA_FILT["KEY"].apply(
    lambda x: generate_presigned_url(bucket_name, object_name=x, expiration=360)
)
```

```python
print("Number of Sounds:", len(DT_HLS_METADATA_FILT))
```

```python
input_files = list(DT_HLS_METADATA_FILT["PRESIGNED_LINK"])  # List of your TS files

# Create a temporary file list for ffmpeg
list_filename = f"{location_}_{unix_datetime}_concat_list.txt"
with open(list_filename, "w", encoding="utf-8") as f:
    for url in input_files:
        f.write(f"file '{url}'\n")

# Define Output File
output_filetype = "flac"
output_file = f"{location_}_{unix_datetime}_sound_file.{output_filetype}"

# Use ffmpeg-python to run the command
ffmpeg_stream = f"ffmpeg -f concat -y -safe 0 -protocol_whitelist file,http,https,tcp,tls,crypto -i {list_filename} -ab 192k {output_file}"
```
